# Regular Graph

[Wiki](https://en.wikipedia.org/wiki/Regular_graph)

**모든 node가 동일한 degree를 갖는 그래프**

이것은 시계열에서의 Graph Shift Operator로써 볼 수 있지.

```python
w=np.zeros((5,5))
for i in range(5):
    for j in range(5):
        if i==j :
            w[i,j] = 0
        elif i-j == 1 : 
            w[i,j] = 1
        elif j-i == 4 :
            w[i,j] = 1
```

```python
w
```

```python
lst = []
for i in range(5):
    for j in range(5):
        if w[i,j] == 1:
            lst.append([i,j])
```

```python
np.array(lst)
```

> 동일한 차수를 갖는다?

::: {.callout-note}
참고로, 무조건 degree matrix가 단위행렬일 필요는 없고 $D = kI$가 성립해도 Regular Graph
:::

```python
d= w.sum(axis=1)
D= np.diag(d)
```

```python
D
```

```python
G = nx.Graph()
```

```python
G.add_edges_from(np.array(lst))
```

```python
plt.figure(figsize=(10, 10)) 
nx.draw_networkx(G, with_labels=True, font_weight='bold', node_color='orange', node_size=1500, font_color='white', font_size=30,width=5)
```

# Non-Euclidean vs Euclidean

## Euclidean

### Ex1) 1D grid

```python
w=np.zeros((5,5))
for i in range(5):
    for j in range(5):
        if i==j :
            w[i,j] = 0
        elif i-j == 1 : 
            w[i,j] = 1
```

```python
lst = []
for i in range(5):
    for j in range(5):
        if w[i,j] == 1:
            lst.append([i,j])
```

```python
d= w.sum(axis=1)
D= np.diag(d)
```

::: {.callout-note}
모든 Degree가 동일한, 특히 단위행렬로 나오는 유클리디안 데이터
:::

```python
D
```

```python
G = nx.Graph()
```

```python
G.add_edges_from(np.array(lst))
```

```python
plt.figure(figsize=(20, 5)) 
nx.draw_networkx(G, with_labels=True, font_weight='bold', node_color='orange', node_size=1500, font_color='white', font_size=30,width=5)
```

### Ex2) 2d grids

```python
w = np.ones((4, 4))
for i in range(4):
    for j in range(4):
        if i==j :
            w[i,j] = 0
```

```python
lst = []
for i in range(4):
    for j in range(4):
        if w[i,j] == 1:
            lst.append([i,j])
```

```python
d= w.sum(axis=1)
D= np.diag(d)
```

::: {.callout-note}
모든 Degree가 동일하여 $D = 3I$로 표현되는 유클리디안 데이터
:::

```python
D
```

```python
G = nx.Graph()
```

```python
G.add_edges_from(np.array(lst))
```

```python
plt.figure(figsize=(10, 10)) 
nx.draw_networkx(G, with_labels=True, font_weight='bold', node_color='orange', node_size=1500, font_color='white', font_size=30,width=5)
```

## Non-Euclidean

### Ex3) Different Weights(2D shape, 1D manifold)

> Weight 같다고 가정하고 그래프 시각화

```python
w=np.zeros((5,5))
for i in range(5):
    for j in range(5):
        if i==j :
            w[i,j] = 0
        elif i!=j: 
            w[i,j] = 1
```

```python
lst = []
for i in range(5):
    for j in range(5):
        if w[i,j] == 1:
            lst.append([i,j])
```

```python
G = nx.Graph()
```

```python
G.add_edges_from(np.array(lst))
```

```python
plt.figure(figsize=(5, 5)) 
nx.draw_networkx(G, with_labels=True, font_weight='bold', node_color='orange', node_size=500, font_color='white', font_size=30,width=5)
```

```python
pi=np.pi
ang=np.linspace(-pi,pi-2*pi/5,5)
r=5+np.cos(np.linspace(0,12*pi,5))
vx=r*np.cos(ang)
vy=r*np.sin(ang)
f1=10*np.sin(np.linspace(0,6*pi,5))
f = f1 + np.random.normal(5)
```

```python
D = np.zeros([5,5])
locations = np.stack([vx, vy],axis=1)
for i in range(5):
    for j in range(i,5):
        D[i,j]=np.linalg.norm(locations[i]-locations[j])
D = D + D.T
```

```python
D
```

::: {.callout-note}
가중치 값이 다 다르게 형성되어 있다. 따라서 $D=kI$형태로도 표현할 수 없어 레귤러 메트릭스의 정의를 충족하지 못하며, 이는 비유클리디안 데이터이다.
:::

```python
fig, ax1 = plt.subplots(1,1,figsize=(5,5),subplot_kw={"projection":"3d"})
ax1.grid(False)
ax1.scatter3D(vx,vy,f,zdir='z',s=50,marker='.',color='gray')
ax1.plot3D(vx,vy,f1,'--k',lw=3)
```
