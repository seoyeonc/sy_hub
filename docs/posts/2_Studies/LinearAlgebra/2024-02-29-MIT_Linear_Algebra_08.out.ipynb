`1` $r=m=n$

-   $R = I$
-   1 solution to $Ax = b$

`2` $r = n<m$

-   $R = \begin{bmatrix} I \\ 0 \end{bmatrix}$
-   0 or 1 solution to $Ax = b$

`3` $r = m<n$, extra rows

-   $R = \begin{bmatrix} I F \end{bmatrix}$
-   no zero row, $\inf$ solution to $Ax = b$

`4` $r<m, r<n$

-   $R = \begin{bmatrix} I & F \\ 0 & 0 \end{bmatrix}$
-   0 or $\inf$ solutions
