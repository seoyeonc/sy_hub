# Tableau

누구나 사용할 수 있는 Data Visulazation Solution

-   장점

    -   drag만으로 figure 만들기 가능
    -   다양한 혀애의 visualization

-   측정값

    -   숫자-액션-집계-차트 순

-   차원

    -   어떻게 나눠서 볼 것인가?

-   불연속형 필드

-   연속형 필드

-   부분전체 분석

command + 1 = show me 단축키

Δ 표시는 테이블 단위로 계산되었다는 표시

-   파이차트보다 트리맵이 좋은 경우

    -   분류 수가 많은 경우 파이차트는 시각적으로 보기 어렵다.

-   분산형 차트

    -   차원은 0 개 이산
    -   측정값이 2~4개인 경우 추천
    -   열 선반과 행 선반에 각각 측정값을 배차하면 자동적으로
        만들어진다.

-   데이터 설명

    -   이상치가 있으면 설명해주기도 함

-   매개 변수 적용

-   이중 축(Dual Axis)

-   결합된 축(Combined Axis)

-   그룹(Group)

    -   차원필드에서 우클릭 만들기 그룹
    -   여러 그룹으로 묶기 가능

-   집합(Set)

    -   교집합 여집합 합집합으로 조건 보기 가능
    -   폴더로 차원보기해서 보고싶른 집합만 폴더로 묶기 가능
    -   in / out으로 표시

-   계층 만들기

    -   계층(hierarchy) 은 화면에서
    -   드릴 다운(Drill-down)해서 값을 세부적으로 찾는 데 유용합니다.

-   총계 만들기

    -   총계(Total)는 각각의 요소들의 총 값을 보여주는 방식으로 범위에
        따라
    -   총합계 및 소계, 그리고 열과 행 기준으로 각각 TOTAL을 표시할 수
        있습니다.

-   지리적 역할 부여하기

-   map 차트 구현하기
